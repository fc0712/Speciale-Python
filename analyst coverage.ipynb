# Analyst Coverage Market cap

```python
import pandas as pd
import joblib
```

```python
df_anal_cov_1Q = pd.read_csv("/content/drive/MyDrive/Cand.fælles/Speciale/analyst_coverage/df_anal_cov_1Q.csv")
df_anal_cov_2Q = pd.read_csv("/content/drive/MyDrive/Cand.fælles/Speciale/analyst_coverage/df_anal_cov_2Q.csv")
df_anal_cov_4Q = pd.read_csv("/content/drive/MyDrive/Cand.fælles/Speciale/analyst_coverage/df_anal_cov_4Q.csv")

Q1_mc = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q1_test_marketcap.pkl")
Q2_mc = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q2_test_marketcap.pkl")
Q4_mc = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q4_test_marketcap.pkl")
```

```python
# Dato format til merging
df_anal_cov_1Q["date"] =  pd.to_datetime(df_anal_cov_1Q["Calc Date"])
df_anal_cov_2Q["date"] =  pd.to_datetime(df_anal_cov_2Q["Calc Date"])
df_anal_cov_4Q["date"] =  pd.to_datetime(df_anal_cov_4Q["Calc Date"])
```

```python
Q1_mc["large_cap"]
```

```python
cap_liste = ["small_cap","mid_cap","large_cap"]
mean_analyst_q1=[]
for cap in cap_liste:
  mean_analyst_q1.append(Q1_mc[cap].reset_index().merge(df_anal_cov_1Q,how="left",left_on=["Forecasttidspunkt","CUSIP"],right_on = ["date","Instrument"])["Number of Analysts"].median())

mean_analyst_q2=[]
for cap in cap_liste:
  mean_analyst_q2.append(Q2_mc[cap].reset_index().merge(df_anal_cov_2Q,how="left",left_on=["Forecasttidspunkt","cusip"],right_on = ["date","Instrument"])["Number of Analysts"].median())

  mean_analyst_q4=[]
for cap in cap_liste:
  mean_analyst_q4.append(Q4_mc[cap].reset_index().merge(df_anal_cov_4Q,how="left",left_on=["Forecasttidspunkt","cusip"],right_on = ["date","Instrument"])["Number of Analysts"].median())
```

```python
dict_df = {"Q1":mean_analyst_q1,"Q2":mean_analyst_q2,"Q4":mean_analyst_q4}
market_analyst = pd.DataFrame(dict_df)
market_analyst.set_index(pd.Series(["small_cap","mid_cap","large_cap"]),inplace=True)
#joblib.dump(market_analyst,"/content/drive/MyDrive/Cand.fælles/Speciale/Python/analyst_cov_marketcap.pkl")
```

```python
Q1_base["FH"] = 1
Q2_base["FH"] = 2
Q4_base["FH"] = 4

```

```python
seaborn_plot = pd.concat([Q1_base,Q2_base,Q4_base])
seaborn_plot["model_error"] = abs(seaborn_plot["Model_predicted"] - seaborn_plot["EPS_actual"])
seaborn_plot["anal_error"] = abs(seaborn_plot["analyst_EPS_mean"] - seaborn_plot["EPS_actual"])

seaborn_plot = seaborn_plot[["model_error","anal_error","FH"]]
```

```python
seaborn_plot_model = seaborn_plot[["FH","model_error"]]
seaborn_plot_model.rename(columns={"model_error":"error"},inplace=True)
seaborn_plot_anal = seaborn_plot[["FH","anal_error"]]
seaborn_plot_anal.rename(columns={"anal_error":"error"},inplace=True)

```

```python
seaborn_plot_model["type"] = "model"
seaborn_plot_anal["type"] = "anal"
```

```python
seaborn_final = pd.concat([seaborn_plot_model,seaborn_plot_anal])
```
