# Metode - dataopdeling

```python
df_train_1 = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q1_train_base.pkl")
df_test_1 = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q1_test_base.pkl")

cutoff_date = "2016-06-30"
validation_start = "2016-09-30"
validation_end = "2018-06-30"
test_start = "2018-09-30"

df_val_1 = df_train_1.loc[validation_start:validation_end,:]
df_train_1 = df_train_1.loc[:cutoff_date,:]
```

```python
train_length = df_train_1.shape[0]
val_length = df_val_1.shape[0]
test_length = df_test_1.shape[0]
total_length = train_length + val_length + test_length

train_pct = round(train_length / total_length,3)
val_pct = round(val_length / total_length,3)
test_pct = round(test_length / total_length,3)

virk_train = df_train_1.CUSIP.nunique()
val_train = df_val_1.CUSIP.nunique()
test_train = df_test_1.CUSIP.nunique()
```

```python
print(virk_train,val_train,test_train)

print(train_pct,val_pct,test_pct)
```
