# Industri

```python
Q1_performance_gen = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q1_performance_generel.pkl")
Q1_performance_ind = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q1_performance_industri.pkl")

Q2_performance_gen = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q2_performance_generel.pkl")
Q2_performance_ind = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q2_performance_industri.pkl")

Q4_performance_gen = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q4_performance_generel.pkl")
Q4_performance_ind = joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q4_performance_industri.pkl")
```

```python
gen_df_1 = pd.concat(Q1_performance_gen).reset_index()
gen_df_1 = gen_df_1[gen_df_1.level_1=="Model"]
ind_df_1 = pd.concat(Q1_performance_ind).reset_index()
ind_df_1 = ind_df_1[ind_df_1.level_1=="Model"]
samlet_df_1 = ind_df_1.merge(gen_df_1,on="level_0",suffixes=("_industry","_generel"))
samlet_df_1.drop(["level_1_industry","level_1_generel"],axis=1,inplace=True)
samlet_df_1.rename(columns={"level_0": "Industri"},inplace=True)
samlet_df_1 = samlet_df_1[["Industri","MAE_industry","MAE_generel"]]
samlet_df_1["Forskel(%)"] = (1-(samlet_df_1["MAE_generel"]/samlet_df_1["MAE_industry"]))*100

gen_df_2 = pd.concat(Q2_performance_gen).reset_index()
gen_df_2 = gen_df_2[gen_df_2.level_1=="Model"]
ind_df_2 = pd.concat(Q2_performance_ind).reset_index()
ind_df_2 = ind_df_2[ind_df_2.level_1=="Model"]
samlet_df_2 = ind_df_2.merge(gen_df_2,on="level_0",suffixes=("_industry","_generel"))
samlet_df_2.drop(["level_1_industry","level_1_generel"],axis=1,inplace=True)
samlet_df_2.rename(columns={"level_0": "Industri"},inplace=True)
samlet_df_2 = samlet_df_2[["Industri","MAE_industry","MAE_generel"]]
samlet_df_2["Forskel(%)"] = (1-(samlet_df_2["MAE_generel"]/samlet_df_2["MAE_industry"]))*100

gen_df_4 = pd.concat(Q4_performance_gen).reset_index()
gen_df_4 = gen_df_4[gen_df_4.level_1=="Model"]
ind_df_4 = pd.concat(Q4_performance_ind).reset_index()
ind_df_4 = ind_df_4[ind_df_4.level_1=="Model"]
samlet_df_4 = ind_df_4.merge(gen_df_4,on="level_0",suffixes=("_industry","_generel"))
samlet_df_4.drop(["level_1_industry","level_1_generel"],axis=1,inplace=True)
samlet_df_4.rename(columns={"level_0": "Industri"},inplace=True)
samlet_df_4 = samlet_df_4[["Industri","MAE_industry","MAE_generel"]]
samlet_df_4["Forskel(%)"] = (1-(samlet_df_4["MAE_generel"]/samlet_df_4["MAE_industry"]))*100

# Concatter dem sammen
samlet_df_alle = pd.concat([samlet_df_1,samlet_df_2,samlet_df_4],axis=1)  
samlet_df_alle.drop("Industri",axis=1,inplace=True)
samlet_df_alle["Industri"] =samlet_df_1["Industri"]
samlet_df_alle.set_index("Industri",inplace=True)

```

# Marketcap tabel 

```python
mcap_tab_1 = pd.concat(joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q1_test_marketcap.pkl"))
mcap_tab_2 = pd.concat(joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q2_test_marketcap.pkl"))
mcap_tab_4 = pd.concat(joblib.load("/content/drive/MyDrive/Cand.fælles/Speciale/Python/Q4_test_marketcap.pkl"))
```

```python
mcap_tab_1.reset_index(level=0,inplace=True)
mcap_tab_2.reset_index(level=0,inplace=True)
mcap_tab_4.reset_index(level=0,inplace=True)

```

```python
def table_values(df):
  IQR_val = df.groupby("level_0")["market_cap"].agg(lambda x: x.quantile(0.75) - x.quantile(0.25))
  median_val = df.groupby("level_0")["market_cap"].median()
  mean_val = df.groupby("level_0")["market_cap"].mean()

  df_val = pd.DataFrame(data=[IQR_val,median_val,mean_val],index=["IQR","MEDIAN","MEAN"])
  return df_val
```

```python
table_val_1 = table_values(mcap_tab_1) / (1*10**6)
table_val_2 = table_values(mcap_tab_2) / (1*10**6)
table_val_4 = table_values(mcap_tab_4) / (1*10**6)
```

```python
pd.concat([table_val_1,table_val_2,table_val_4],axis=1).round(2)
```
